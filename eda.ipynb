```python
f_sample = np.fft.fft(sample)
```

```python
r_f_sample = np.real(f_sample)
c_f_sample = np.imag(f_sample)
timestep = 1/250
freq = np.fft.fftfreq(r_f_sample[0].size, d=timestep)
```
